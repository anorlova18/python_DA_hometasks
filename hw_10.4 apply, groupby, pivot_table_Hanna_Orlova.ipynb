3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = pd.pivot_table(df,values= 'Invoice ID', index = 'Branch', columns = 'Product line', aggfunc = 'count')
```

Гіпотеза1. Яка категорія має найбільшу популярність (максимум в одному філіалі)?

```python
invoices_by_category.max(axis=0).idxmax()
```

- Використовувати як "магніт" для клієнтів
- Робити акції навколо цієї категорії

Гіпотеза 2. У якому філіалі найбільший розрив між найкращою і найгіршою категорією?

```python
(invoices_by_category.max(axis=1) - invoices_by_category.min(axis=1)).idxmax()
```

- Нерівномірний попит
- або розвивати слабкі категорії, або фокус на сильних

Гіпотеза 3. Яка категорія має найменший розкид між філіалами (найбільш передбачувана)?

```python
(invoices_by_category.max(axis=0) - invoices_by_category.min(axis=0)).idxmin()
```

- Найбільш стабільний товар
- добре підходить для прогнозування та планування запасів

Гіпотеза 4. Яка категорія має найменший загальний попит?

```python
invoices_by_category.sum(axis=0).idxmin()
```

Треба:
- або зменшити інвестиції
