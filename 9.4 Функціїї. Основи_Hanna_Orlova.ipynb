2. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `calculation` —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ 4 –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É int - `a`, `b`, `c`, `d` —ñ –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç–∞–∫–æ—ó —Ñ–æ—Ä–º—É–ª–∏: $ (a + b - c) * d $.

  –í–∏–∫–ª–∏—á—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –Ω–∞–±–æ—Ä—ñ –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤ —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω:
  ```
  a = 16
  b = 4
  c = 9
  d = 3
  ```

````python
def calculation(a: int, b: int, c: int, d: int):
  return (a + b - c)*d
result = calculation(16,4,9,3)
print(result)
````

4. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `multipy_numbers`, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –¥–≤–∞ —á–∏—Å–ª–∞ —è–∫ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —ñ –ø–æ–≤–µ—Ä—Ç–∞—î —ó—Ö –¥–æ–±—É—Ç–æ–∫.

  –í–∏–∫–ª–∏—á—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `multipy_numbers` –Ω–∞ –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª
- 10 —ñ 2
- 7 —ñ 8
- 1024 —ñ 999.
  
  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–æ–∂–Ω–æ–≥–æ –≤–∏–∫–ª–∏–∫—É.
  –ü—Ä–∏ –≤–∏–∫–æ–Ω–∞–Ω–Ω—ñ —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ü–∏–∫–ª—É - –æ—Ç—Ä–∏–º–∞—î—Ç–µ 2—Ö –±–∞–ª—ñ–≤ –∑–∞ –Ω—å–æ–≥–æ :)

````python
def multipy_numbers(x: int, y: int):
  return x * y
print(multipy_numbers(10, 2))
print(multipy_numbers(7, 8))
print(multipy_numbers(1024, 999))
````

````python
def multipy_numbers(x: int, y: int):
  return x * y
nabor = [(10, 2), (7, 8), (1024, 999)]
for x, y in nabor:
     result = multipy_numbers(x, y)
     print(result)
````

5. **–î—ñ–ª–µ–Ω–Ω—è –∑ –ø–µ—Ä–µ–≤—ñ—Ä–∫–æ—é —Ç–∏–ø—ñ–≤.**  

  1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `int_division` —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É `int` - `a` —ñ `b`, —ñ –≤–∏–∫–æ–Ω—É—î –¥—ñ–ª–µ–Ω–Ω—è –ø–µ—Ä—à–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –Ω–∞ –¥—Ä—É–≥–∏–π: `a/b`. –ü—Ä–∏ —Ü—å–æ–º—É —Ñ—É–Ω–∫—Ü—ñ—è –ø–µ—Ä–µ–¥ –¥—ñ–ª–µ–Ω–Ω—è–º –ø–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ –¥—Ä—É–≥–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ –¥–æ—Ä—ñ–≤–Ω—é—î 0. –Ø–∫—â–æ –¥—Ä—É–≥–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ 0, —Ñ—É–∫–Ω—Ü—ñ—è –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥—ñ–ª–µ–Ω–Ω—è, —ñ–Ω–∞–∫—à–µ - –∑–∞–º—ñ—Å—Ç—å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –¥—ñ–ª–µ–Ω–Ω—è, –≤–µ—Ä—Ç–∞—î –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –Ω–∞ –µ–∫—Ä–∞–Ω "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0".

    –Ø–∫—â–æ —Ç–∏–ø–∏, –Ω–∞–¥–∞–Ω—ñ —É —Ñ—É–Ω–∫—Ü—ñ—é, –Ω–µ `int` - —Ñ—É–Ω–∫—Ü—ñ—è –≤–∏–≤–æ–¥–∏—Ç—å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.".

  2. –ó–∞–ø—É—Å—Ç—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª:
- 12 —ñ 6
- 20 —ñ 0
- 112 —ñ 10
- 99 —ñ 0
- 11 i 1.1

    —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: " - —ñ –¥–∞–ª—ñ —Ç–µ, —â–æ –ø–æ–≤–µ—Ä–Ω—É–ª–∞ —Ñ—É–Ω–∫—Ü—ñ—è.

  3. –î–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: "–ß–æ–º—É –≤ –¥–µ—è–∫–∏—Ö –≤–∏–ø–∞–¥–∫–∞—Ö —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä–Ω—É–ª–∞ None?"

````python
def int_division(a, b):
  # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ –Ω–∞ —Ç–∏–ø int
    if type(a) is not int or type(b) is not int:
      print("–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.")
      return
# –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0
    if b == 0:
      print("–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0")
      return
# –ü—ñ—Å–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–æ–∫ –≤–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    return(a/b)
# –ù–∞–±—ñ—Ä –¥–∞–Ω–Ω–∏—Ö
nabor_1 = [(12, 6), (20, 0), (112, 10), (99, 0), (11, 1.1)]
# –¶–∏–∫–ª
for a, b in nabor_1:
    result = int_division(a,b)
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: {result}")
````

````python
def int_division(a, b):
    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ, —á–∏ –æ–±–∏–¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ –º–∞—é—Ç—å —Ç–∏–ø int
    if type(a) is not int or type(b) is not int:
        print("–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.")
        return

    # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ –Ω—É–ª—å
    if b == 0:
        print("–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0")
        return

    # –Ø–∫—â–æ –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –ø—Ä–æ–π–¥–µ–Ω—ñ, –ø–æ–≤–µ—Ä—Ç–∞—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    return a / b

# –ù–∞–±—ñ—Ä —Ç–µ—Å—Ç–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö
test_pairs = [
    (12, 6),
    (20, 0),
    (112, 10),
    (99, 0),
    (11, 1.1)
]

# –ó–∞–ø—É—Å–∫ —Ñ—É–Ω–∫—Ü—ñ—ó –≤ —Ü–∏–∫–ª—ñ
for a, b in test_pairs:
    result = int_division(a, b)
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: {result}")

#–£ Python –±—É–¥—å-—è–∫–∞ —Ñ—É–Ω–∫—Ü—ñ—è, —è–∫–∞ –∑–∞–≤–µ—Ä—à—É—î —Å–≤–æ—é —Ä–æ–±–æ—Ç—É –±–µ–∑ —è–≤–Ω–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ return –∑ —è–∫–∏–º–æ—Å—å –∑–Ω–∞—á–µ–Ω–Ω—è–º, –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –ø–æ–≤–µ—Ä—Ç–∞—î None.
#–£ –Ω–∞—à–æ–º—É –≤–∏–ø–∞–¥–∫—É —Ü–µ —Å—Ç–∞—î—Ç—å—Å—è –¥–≤–∞ —Ä–∞–∑–∏:
#–ü–æ–º–∏–ª–∫–∞ —Ç–∏–ø—É: –ø—Ä–∏ –ø–µ—Ä–µ–≤—ñ—Ä—Ü—ñ type(a) is not int..., –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è print, –∞ –¥–∞–ª—ñ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–≤–µ—Ä—à—É—î—Ç—å—Å—è (–º–∏ –¥–æ–¥–∞–ª–∏ –ø—É—Å—Ç–∏–π return –∞–±–æ –≤–æ–Ω–∞ –ø—Ä–æ—Å—Ç–æ –¥–æ—Ö–æ–¥–∏—Ç—å –¥–æ –∫—ñ–Ω—Ü—è –±–ª–æ–∫—É if).
#–î—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0: –ö–æ–ª–∏ b == 0, –≤–∏–≤–æ–¥–∏—Ç—å—Å—è –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è, —ñ —Ñ—É–Ω–∫—Ü—ñ—è –∑–Ω–æ–≤—É –∑–∞–≤–µ—Ä—à—É—î—Ç—å—Å—è –±–µ–∑ –ø–æ–≤–µ—Ä–Ω–µ–Ω–Ω—è —á–∏—Å–ª–æ–≤–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É.
#–û—Å–∫—ñ–ª—å–∫–∏ –≤ —Ü–∏–∫–ª—ñ —É print(f"... {result}"), –∞ –≤ –∑–º—ñ–Ω–Ω—É result –ø–æ—Ç—Ä–∞–ø–∏–ª–æ "–Ω—ñ—â–æ" (—Ç–æ–±—Ç–æ None), –º–∏ –±–∞—á–∏–º–æ —Ü–µ —Å–ª–æ–≤–æ –Ω–∞ –µ–∫—Ä–∞–Ω—ñ.
````
